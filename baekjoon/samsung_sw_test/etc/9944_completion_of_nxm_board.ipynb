# NxM 보드 완주하기

## 문제 링크

* https://www.acmicpc.net/problem/9944

## 코드

```python
def dfs(board, n, m, n_empty, x, y, visited=set()):
    def move(x, y, dx, dy, n, m, visited):
        WALL, new_visited = '*', set()
        x, y = x + dx, y + dy
        while 0 <= x < n and 0 <= y < m \
        and board[x][y] != WALL and (x, y) not in visited:
            new_visited.add((x, y))
            x, y = x + dx, y + dy
        x, y = x - dx, y - dy
            
        return x, y, new_visited
        
    if len(visited) == n_empty:
        return 0
    
    min_count = 1000
    for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
        next_x, next_y, new_visited = move(x, y, dx, dy, n, m, visited)
        visited.update(new_visited)
        if not(x == next_x and y == next_y):
            count = 1 + dfs(board, n, m, n_empty, next_x, next_y, visited)
            min_count = min(count, min_count)
        visited.difference_update(new_visited)
        
    return min_count

def solution(board, n, m):
    EMPTY = '.'
    ball_positions = []
    n_empty = 0
    for x in range(n):
        for y in range(m):
            if board[x][y] == EMPTY:
                ball_positions.append((x, y))
                n_empty += 1
    
    MAX_COUNT = 1000
    min_count = MAX_COUNT
    for x, y in ball_positions:
        count = dfs(board, n, m, n_empty, x, y, set([(x, y)]))
        min_count = min(count, min_count)
    return min_count if min_count != MAX_COUNT else -1
```

### 예제 입력 및 출력

```python
lines = [
    '7 7\n', #10
    '.......\n',
    '.......\n',
    '.......\n',
    '*******\n',
    '.......\n',
    '.......\n',
    '.......\n',
]
```

```python
import sys

lines = sys.stdin.readlines()
lengths, boards, i = [], [], 0
while i < len(lines):
    N, M = map(int, lines[i].split())
    lengths.append((N, M))
    board = []
    for j in range(N):
        board.append(list(lines[i+1+j][:-1]))
    boards.append(board)
    i = i + N + 1

for i, ((n, m), board) in enumerate(zip(lengths, boards), start=1):
    print('Case %d: %d' % (i, solution(board, n, m)))
```

#### 예제 1

```python
N, M = 5, 5
board = [
    list('**...'),
    list('.....'),
    list('....*'),
    list('..*..'),
    list('.....')
]
solution(board, N, M)
```
